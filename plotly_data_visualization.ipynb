```python
import plotly.express as px #installing important libraries
import pandas as pd
```

#**Bar chart**

```python
#Read data into DataFrame
df = pd.read_csv("/content/used cars.csv")
```

```python
#checking the first 5 dataset content
df.head()
```

```python
#creating a bar cahrt using plotly
fig=px.bar(df, x="Brand", y="Cars Listings", color="Cars Listings", height=400)
#adding a title and labels to the x and y axis
fig.update_layout(title_text="Cars Listings by Brand", xaxis_title="Brand",
                  yaxis_title="Cars Listings")
```

```python
#customizing color cell and adding a color bar
fig.update_traces(marker=dict(color='red', opacity=0.7,
                              line=dict(color='red', width=1.5)))
fig.update_layout(coloraxis_colorbar=dict(title="Cars Listings"))
```

# **Pie chart**

```python
#Converting the dictionary to a pandas dataframe,
#it will be used to create a pie chart
data = pd.DataFrame(df)
```

```python
data.head()
```

```python
#Create a pie chart using Plotly Express, specifying the data frame,
#the column for the values, the column for the names, and the color
#for each slice.
fig = px.pie(df, values='Cars Listings', names='Brand',color='Cars Listings',
             title= 'Car Listings Brand')
fig.show()
```

```python
#Update the text position of the labels to be inside each slice
#and show the percent and label information
fig.update_traces(textposition='inside', textinfo='percent+label')
```

```python
#Update the layout to set a minimum size for the text
# and to hide any overlapping text

fig.update_layout(uniformtext_minsize=14, uniformtext_mode='hide')
```

# **scatter plot**

```python
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
dfscatter = pd.read_csv('/content/scatter_data.csv')
```

```python
dfscatter.head()
```

```python
#Create a trace using the go.Scatter() function. This will create a scatte
trace = go.Scatter(x=dfscatter['Area (ft.)'], y=dfscatter['Price'],
                   mode='markers',
                   marker=dict(color=dfscatter['Building Type']))

#Create a data list. Here, the trace object created in
#the previous step is added to a list called data.
data= [trace]

#Create a layout object using the go.Layout() function.
#This will create a layout for the chart, including the title,
#X and Y axis labels.
layout = go.Layout(title='Year of Sale vs. Price',
                   xaxis=dict(title='Year of sale'),
                   yaxis=dict(title='Price'))

#Create a figure using the go.Figure() function.
#This combines the data and layout objects to create a complete figure.
fig = go.Figure(data=data, layout=layout)

fig.show()

#Show the plot using pyo.plot() function. This function takes the fig
#object as an argument and displays the plot in the default web browser.
pyo.plot(fig)
```
